```python
import numpy as np 
n_before = np.random.rand(3,4)
print(n_before)
n_tensor = torch.from_numpy(n_before)
print(n_tensor)
n_after = n_tensor.numpy ()
print(n_after)
```

```python
x = torch.tensor([1,2,4])
y = torch.tensor([1,2,4])
print(x+y)
print(torch.add(x,y))
print(torch.mul(x,y))
print(torch.div(x,y))
print(torch.true_divide(x,y))
```

```python
# Inplace operation (fjd_)
t = torch.ones(3)
print(t+x)
print(t.add_(x)) # It adds and replaces the content
print(t+x) # t+ = x == this is also inplace
```

```python
print(x.pow(2))
print(x**2)
```

```python
print (x<y)
print (y>0)
print (x==y)
print (x>=y)
```

```python
x1 = torch.rand((2,5))
x2 = torch.rand((5,3))
print(torch.mm(x1,x2))
print(x1.mm(x2))
print(torch.dot(x,y))
print(x.dot(y))
```

```python
# Batch matrix multiplicatin
batch =32
n=10
m=20
p=30

tensor1= torch.rand(batch,n,m)
tensor2= torch.rand(batch,m,p)
bmm = torch.bmm(tensor1,tensor2)
```

```python
# Broadcasting
x1 = torch.rand((3,5))
x2 = torch.rand ((1,5))

print(x1)
print(x2)
print(x2*x1)
```

```python
# Other useful operations
sum_x  =torch.sum(x1)
sum_x_dim0  =torch.sum(x1, dim=0)
sum_x_dim1  =torch.sum(x1, dim=1)
print(sum_x)
print(sum_x_dim0)
print(sum_x_dim1)
```

```python
max_x = torch.max(x1)
max_x_dim0 = torch.max(x1, dim=0)
max_x_dim1 = torch.max(x1, dim=1)
print(max_x)
print(max_x_dim0)
print(max_x_dim1)

val, ind = torch.max(x1, dim=0)
print(val)
print(ind)
```

```python
# Can use dimension attribute for rows and column
print(torch.abs(x1))
print(torch.argmax(x1))
print(torch.argmin(x1))
print(torch.mean(x1))
print(torch.mean(x1, dim=0))
print(torch.std(x1))
print(torch.var(x1))
```

```python
print (torch.sort(x1, descending=True, dim=0))
```

```python
# clamp
z = torch.tensor([-1,0,4,5,10,7,-6,22,33])
print(torch.clamp(z, min=0, max=10,)) # Basically like RELU function
# anything below zero will be set to 0 and anything above 10 will be set to 10
```

```python
bool_val = torch.tensor([1,0,0,1,1,0], dtype=torch.bool)
print(bool_val)
print(torch.any(bool_val))
print(torch.all(bool_val))
```
